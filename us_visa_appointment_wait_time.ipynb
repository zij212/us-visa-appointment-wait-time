```python
location_list = eval(source)
```

```python
google_api_key = 'replace_with_your_kry'
```

```python
for embassy in location_list:
  # change N'Djamena to NDjamena before adding it to the url
  city = re.sub(f'[{string.punctuation}]', '', embassy['value'])
  try:
    response = requests.get(f'https://maps.googleapis.com/maps/api/geocode/json?address={city}&key={google_api_key}')
    embassy['coordinates'] = response.json()['results'][0]['geometry']['location']
  except Exception as e:
    print(city, e)
```

```python
with open('locations.json', 'w') as f:
  json.dump(location_list, f)
```

```python
with open('./locations.json', 'r') as f:
  location_list = json.loads(f.read())
```

```python
%%time
for embassy in location_list:
  code = embassy['code']
  try:
    response = requests.get(f"https://travel.state.gov/content/travel/resources/database/database.getVisaWaitTimes.html?cid={code}&aid=VisaWaitTimesHomePage")
    visitor_visa, student_visa, all_other_nonimmigrant_visa, unknown_category_visa = response.text.replace('\r', '').replace('\n', '').split(',')
    embassy['visitor_visa'] = visitor_visa
    embassy['student_visa'] = student_visa
    embassy['all_other_nonimmigrant_visa'] = all_other_nonimmigrant_visa
    embassy['unknown_category_visa'] = unknown_category_visa
  except Exception as e:
    print(embassy, e)
  # break
```

```python
df = pd.DataFrame.from_records(location_list)
df.rename(columns={'code': 'embassy_code', 'value': 'embassy_city', 'coordinates': 'city_coordinates'}, inplace=True)
df.to_csv('us_visa_appointment_wait_time.csv', index=False)
```

```python
def waiting_day_to_color(number: int) -> str:
  if number <= 30:
    return 'green'
  elif number <= 100:
    return 'orange'
  elif number <= 400:
    return 'red'
  else:
    return 'black'

    
def waiting_time_str_to_int(wait_time: str) -> int: 
  if 'Closed' in wait_time:
    return 1000
  elif 'Same Day' == wait_time:
    return 0
  elif 'Emergency Appointments Only' == wait_time:
    return 999
  else:
    return int(re.findall(r'[0-9]+', wait_time)[0]) 
```

```python
m = folium.Map(location = (25.901271, -3.470986), zoom_start=3)


for embassy in df.itertuples():
  visitor_visa_wait_time = waiting_time_str_to_int(embassy.visitor_visa)
  marker_color = waiting_day_to_color(visitor_visa_wait_time)
  city = re.sub(f'[{string.punctuation}]', '', embassy.embassy_city)

  folium.Marker(
    [embassy.city_coordinates['lat'], embassy.city_coordinates['lng']], 
    icon=folium.Icon(color=marker_color, icon='circle', prefix='fa'), 
    # tooltip=f"{city}: {embassy.visitor_visa}", 
    popup=f"{city}: {embassy.visitor_visa}"
).add_to(m)
  # break
display(m)
```
